```python
import pickle
from matplotlib import pyplot as plt
import os
import cv2

with open('trainnvalPaper.pickle','rb') as f:
    im = pickle.load(f)
with open('trainnval2.txt','r',encoding='utf-8') as f:
    lines = f.readlines()
for i in range(7000,8000):
    line = lines[i]
    lineSplit = line.strip().split(' ')
    img = cv2.imread(lineSplit[0],0)
    plt.subplot(2,1,1);
    plt.imshow(im[:,:,i],cmap='gray')
    plt.title(str(i+1))
    plt.subplot(2,1,2);plt.imshow(img,cmap='gray')
    plt.show()
```
